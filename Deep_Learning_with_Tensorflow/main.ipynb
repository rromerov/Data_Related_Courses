# ¿Qué es un Tensor?
Los datos que son pasados entre las operaciones son los tensores. En consecuencia un Tensor es un arreglo multidimensional, puede ser de 0 dimensiones (valor escalar), unidimensional (un vector) o bidimensional (como una matriz)

La librería tensorflow permite redimensionar los datos a la forma en la que los queramos, muy útil con imágenes, podemos ver a las imágenes con color como otra dimensión (3D).

# ¿Por qué usar deep learning con tensorflow? 

* Soporte extensivo para deep learning.

* Funciones matemáticas para redes neuronales.

* Optimizadores autodiferenciales y optimizadores de primera clase.


# Actualización TF se encuentra en la versión 2.x

* TF 2.x, Keras es la API de alto nivel predeterminada para Tensorflow.



El siguiente código está hecho en Tensorflow 2

````python
import tensorflow as tf
import numpy as np

#inicializamos 2 tensores
a=tf.constant(np.array([1.,2.,3.]))
b=tf.constant(np.array([4.,5.,6.]))

#después calculamos el producto punto del tensor y asignamos el resultado a C.
c=tf.tensordot(a,b,1)

#mostramos el resultado
output=c.numpy()
output
````

Con Tensorflow el código va de la siguiente manera:



```
import tensorflow as tf
import numpy as np

a=tf.constant(np.array([1.,2.,3.]))
b=tf.constant(np.array([4.,5.,6.]))
c=tf.tensordot(a,b,1)

session=tf.Session()
output=session.run(c)
session.close()
```

Con Tensorflow 2.x ya no es necesario iniciar la sesión.
